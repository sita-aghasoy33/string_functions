```python
words = 'The life is a box full of chocolates.'
list_of_words = words.split(' ')

def most_repated_letters(word_list):
  temp_list = []
  """
  this function returns the most repeated letters in the given word and the 
  number of them in a dictionary type object.
  Suitable for converting to DataFrame.
  """
  # for each word in list
  for word in word_list:
    # create a temporary dict for each word
    letters_dict = {}

    # if word is at least one element long
    if len(word) >= 1:

      # for each letter in word
      for letter in word:
          # create a key named as letter itself and assign its frequency to it.
          letters_dict[letter] = word.count(letter)

      # create a list of most frequent letter for each word
      temp_letter_list = [pair[0] for pair in letters_dict.items() 
                            if pair[1] == max(list(letters_dict.values()))]

      # create a final list with tuple for each word consisting of:
      #   - worf itself;
      #   - list of most frequent letters;
      #   - and number of frequency.
      temp_list.append((word,
                        temp_letter_list,
                        max(list(letters_dict.values()))))
  # create an empty dictionary with keys
  final_dict = {'word':[], 'most repated letters':[], 'number of letters':[]}

  # for each tuple in temp_list
  #    - assign first element of tuple (word itself) to 'word';
  #    - assign second element of tuple (most frequent letters) to 'most repated letters';
  #    - assign third element of tuple (the number of letters) to 'number of letters';
  for tup in temp_list:
    final_dict['word'].append(tup[0])
    final_dict['most repated letters'].append(tup[1])
    final_dict['number of letters'].append(tup[2])

  return final_dict

# converting dictionary to a DataFrame object
pd.DataFrame(most_repated_letters(list_of_words)).sort_values(by='number of letters', ascending = False)
```

```python
#@title Write a function called word frequencies(mylist) that accepts a list of strings called mylist and returns a dictionary where the keys are the words from mylist and the values are the number of times that word appears in mylist:
mystring = "aaaaabbbbcaccaaaadde"
"""
This function returns the frequency of elements in given string.
"""
def frequencies(string):
  # create an empty dictionry
  element_freq = {}

  # for each character in string
  for i in string:
      # create key named as character itself,...
      # ...and assign the frequency to it
      element_freq[i] = string.count(i)
  return element_freq

frequencies(mystring)
```

```python
#@title Take an input from user (at least 2 sentences). Then, print the hard and soft vowels in the given text. Reference: https://www.azerbaijans.com/content_939_en.html { form-width: "500px" }
def vowel_extraction(text):
    vowels = {}
    hard_vowels = []
    soft_vowels = []

    # for each character in text
    for i in text:
        # if it is hard vowel...
        if i.lower() in ['a', 'ı', 'o', 'u']:
            # ...append it to 'hard_vowels' list
            hard_vowels.append(i)
         # if it is soft vowel...
        elif i.lower() in ['ə', 'e', 'ö', 'ü', 'i']:
            # ...append it to 'soft_vowels' list
            soft_vowels.append(i)
    # create a dictionary and assign lists to corresponding keys
    vowels['hard vowels'] = hard_vowels
    vowels['soft vowels'] = soft_vowels
    print(vowels)
```
